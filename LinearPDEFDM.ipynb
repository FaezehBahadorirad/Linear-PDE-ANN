```julia
using Plots
```

```julia
# Parameters
L = 1.0
Î± = 0.01
u0 = 100.0
uL = 0.0
n = 100  # Number of grid points
h = L / (n - 1)  # Grid spacing
```

```julia
# Grid
x = range(0, stop=L, length=n)

# Initialize u
u = zeros(n)
u[1] = u0
u[end] = uL

# Construct the finite difference matrix
A = zeros(n, n)
for i in 2:(n-1)
    A[i, i-1] = 1.0 / h^2
    A[i, i] = -2.0 / h^2 - Î±
    A[i, i+1] = 1.0 / h^2
end

# Apply boundary conditions
A[1, 1] = 1
A[n, n] = 1

# Solve the linear system
u = A \ u

# Plotting
y_dummy = zeros(n)  # Dummy y-axis
surface(x, y_dummy, u, title="3D FDM Solution", xlabel="x", ylabel="y", zlabel="u(x)", label="FDM solution")
```
